## Locked Doors Env (CIFAR10)

```python
env = gym.make('CIFARLockedDoorsEnvTest-v2')
```

```python
obs = env.reset()
{k: v.shape for k, v in obs.items()}
```

```python
image = (np.moveaxis(obs['descriptor'], 0, 2))
# CIFAR-10 normalization
mu = np.array((0.485, 0.456, 0.406))
std = np.array((0.229, 0.224, 0.225))
```

```python
plt.imshow(image)
plt.figure()
plt.imshow(image * std + mu)
```
